```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data from the problem
income = [30, 50, 70, 80, 100]  # x1
websites = [0, 1, 1, 2, 1]       # x2
donate = [0, 1, 0, 1, 1]         # y

# Converting the lists to numpy arrays for easier indexing
income = np.array(income)
websites = np.array(websites)
donate = np.array(donate)

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(income[donate == 1], websites[donate == 1], c='blue', label='Donate = Yes', marker='o')
plt.scatter(income[donate == 0], websites[donate == 0], c='red', label='Donate = No', marker='x')

# Labeling the axes
plt.xlabel('Income (thousands $)')
plt.ylabel('Number of websites')
plt.title('Scatter Plot of Donation Data')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Sample data
income = np.array([30, 50, 70, 80, 100])
websites = np.array([0, 1, 1, 2, 1])
y_train = np.array([0, 1, 0, 1, 1])  # The actual labels for the data

# Sample weights and bias
m1 = 0.4 # weight for income
m2 = 30 # weight for websites
b = -45   # bias

# Create a range of values for income to calculate the decision boundary
x_values = np.linspace(start=min(income), stop=max(income), num=300)

# Calculate the corresponding websites values using the equation of the line
# 0 = m1 * income + m2 * websites + b
# websites = (-m1/m2) * income - b/m2
y_values = (-m1/m2) * x_values - b/m2

# Plot the original data points
plt.scatter(income[donate == 1], websites[donate == 1], c='blue', label='Donate = Yes', marker='o')
plt.scatter(income[donate == 0], websites[donate == 0], c='red', label='Donate = No', marker='x')
# Plot the decision boundary
plt.plot(x_values, y_values, label='Decision Boundary', color='green')

# Label the axes and show the plot
plt.xlabel('Income (thousands $)')
plt.ylabel('Number of Websites')
plt.title('Decision Boundary Plot')
plt.legend()
plt.grid(True)
plt.show()
```
